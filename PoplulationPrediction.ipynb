```python
def abline(slope, intercept):
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--', color="red")
```

```python
def predict_population_2025(country):
    # Clean the input country name (remove leading/trailing whitespace, convert to lowercase)
    country_cleaned = country.strip().lower()
    
    # Find the row corresponding to the country
    country_data = data[data['Country'].str.strip().str.lower() == country_cleaned]
    
    if country_data.empty:
        print(f"Country '{country}' not found in the dataset.")
        return
    
    # Extract Population 2023, Population 2024, Density (/km2), and Growth Rate for the country
    population_2023 = country_data['Population 2023'].values[0]
    population_2024 = country_data['Population 2024'].values[0]
    density = country_data['Density (/km2)'].values[0]
    growth_rate = country_data['Growth Rate'].values[0]
    
    # Predict Population 2025 based on the features
    population_2025_pred = lr.predict([[population_2023, population_2024, density, growth_rate]])
    print(f"Predicted Population for {country} in 2025: {population_2025_pred[0]:,.0f}")

    plt.figure(figsize=(8, 6))
    plt.bar(['2024', '2025'], [population_2024, population_2025_pred[0]], color=['blue', 'orange'])
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.title(f'Population for {country.title()} in 2024 and Predicted 2025. In 2024')
    plt.show()

while True:
    try:
        country = input("Enter the name of the country: ")
        predict_population_2025(country)
        break
    except ValueError:
        print("Country not found or invalid input. Please try again.")
```
